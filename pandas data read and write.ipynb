```python
pd.read_table('ch_01.txt',sep ='\d')
```

```python
pd.read_table('ch_01.txt',sep ='\D')
```

```python
pd.read_table('ch_01.txt',sep ='\s')
```

```python
pd.read_table('ch_01.txt',sep ='\S')
```

```python
pd.read_table('ch_01.txt',sep ='\n')
```
