```python
df=pd.read_csv('Melbourne_housing_FULL.csv')
df.head()
```

```python
important_columns=['Suburb', 'Rooms', 'Type', 'Method', 'SellerG','Regionname','Propertycount',
        'Distance', 'CouncilArea','Bedroom2', 'Bathroom', 'Car',
       'Landsize', 'BuildingArea','Price']
df=df[important_columns]
df.head()
```

```python
df.isna().sum()
```

```python
set_zero=['Propertycount','Distance','Bedroom2','Bathroom','Car']
```

```python
df[set_zero]=df[set_zero].fillna(0)
```

```python
df.isna().sum()
```

```python
df['Landsize']=df['Landsize'].fillna(df.Landsize.mean())
```

```python
df['BuildingArea']=df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df=pd.get_dummies(df,drop_first=True)
df
```

```python
x=df.drop('Price',axis=1)
x
```

```python
y=df['Price']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.linear_model import Lasso
model=Lasso(alpha=50,max_iter=100,tol=0.1)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.linear_model import Ridge
model=Ridge(alpha=50,max_iter=100,tol=0.1)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```
