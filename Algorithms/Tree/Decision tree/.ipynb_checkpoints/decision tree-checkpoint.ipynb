```python
import pandas as pd
df=pd.read_csv('salaries.csv')
df.head()
```

```python
x_df=df.drop(['salary_more_then_100k'],axis=1)
x_df
```

```python
x=pd.get_dummies(x_df)
x.head()
```

```python
y=df['salary_more_then_100k']
y.head()
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[0,0,1,0,0,1,1,0]])
```

# Excercise- Titanic Survival Prediction

```python
df1=pd.read_csv('titanic.csv')
df1.head()
```

```python
df1.shape
```

```python
x_df1=df1.drop(['PassengerId','Survived','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis=1)
x_df1
```

```python
y=df1['Survived']
y
```

```python
x=pd.get_dummies(x_df1)
x
```

```python
x.isnull().sum()
```

```python
import numpy as np
```

```python

x['Age'].fillna((x['Age'].mean()),inplace=True)
x
```

```python
x.head(20)
```

```python
x.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_test
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
