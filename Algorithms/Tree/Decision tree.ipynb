## What is Decision tree?

Decision tree is the most powerful and popular tool for classification and prediction. A Decision tree is a flowchart like tree structure, where each internal node denotes a test on an attribute, each branch represents an outcome of the test, and each leaf node (terminal node) holds a class label.


* Supervised Learning
* Works for both classification and regression
* Foundation of Random Forests
* Attractive Because of interpretability

Decision tree works by:
* Split based on set impurity criteria
* Stopping criteria


Advantages:

* Simple to understand and to interpret.Trees can be visualised.
* Require little data preparation.
* Able to handle both numerical and categorical data.
* Possible to validate a model using statistical tests.
* Perform well even if its assumptions are somewhat violated by the true model from which the data were generated.


Disadvantages:

* Overfitting
* Can be unstable.
* Can't guarantee to return the globally optimal decision tree.

## Training and Visualization Decision tree

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sys
from sklearn import tree
```

```python
%matplotlib inline
```

```python
x=[[0,0],[1,2]]
y=[0,1]
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x,y)
```

```python
model.predict([[2.,2.]])
```

```python
model.predict_proba([[2.,2.]])
```

```python
model.predict([[0.4,1.2]])
```

```python
model.predict_proba([[0.4,1.2]])
```

```python
model.predict_proba([[0,0.2]])
```

DecisionTreeClassifier is capable of both binary(where the labels are[-1,1]) classification and multiclass(where the labels are[0,k-1]) classification
## Applying to Iris Dataset.

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.data[0:5]
```

```python
iris.feature_names
```

```python
x=iris.data[:,2:]
x
```

```python
y=iris.target
y
```

```python
model=tree.DecisionTreeClassifier(random_state=42)
model.fit(x,y)
```

## Export_graphviz
