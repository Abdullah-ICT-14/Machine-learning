# which one should you use ? (pyplot or matplotlib oo method)
* when ploting something quickly okay to use the pyplot method
* when ploting something more advanced use the oo method

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease['age']>50]
```

```python
over_50.head()
```

```python
# pyplot method
over_50.plot(kind='scatter',x='age',y='chol',c='target',figsize=(10,5));
```

```python
# oo method
fig,ax=plt.subplots(figsize=(10,5))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);
#ax.set_xlim(45,100)
```

```python
over_50
```

```python
# oo method and customizing plot
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])

# customizing plot
ax.set(title='ploting by using object oriented method',xlabel='age',ylabel='cholosterol')

# add a legend
ax.legend(*scatter.legend_elements(),title='target');

# add a horizontal line
ax.axhline(over_50['chol'].mean())
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))


s=ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
ax0.set(title='Heart diseases',xlabel='Age',ylabel='cholorestrol');
ax0.legend(*s.legend_elements(),title='target')
ax0.axhline(over_50['chol'].mean());

s=ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'])
ax1.set(title='Heart diseases',xlabel='Age',ylabel='blood pressure');
ax1.legend(*s.legend_elements(),title='target')
ax1.axhline(over_50['thalach'].mean());

# added title to the figure

fig.suptitle('Heart diseases Analysis',fontsize=12,fontweight='bold')
```

```python
plt.style.use('seaborn-whitegrid')
# oo method and customizing plot
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'],cmap='winter')

# customizing plot
ax.set(title='ploting by using object oriented method',xlabel='age',ylabel='cholosterol')

# add a legend
ax.legend(*scatter.legend_elements(),title='target');

# add a horizontal line
ax.axhline(over_50['chol'].mean())
```

```python
# customizing the y and x axis limitations

fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))

#add data to ax0

s=ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'],cmap='winter')

# customizing ax0
ax0.set(title='Heart diseases',xlabel='Age',ylabel='cholorestrol');

# change the x axis limits
ax0.set_xlim([50,80])

# adding legend to ax0
ax0.legend(*s.legend_elements(),title='target')

#adding a horizontal mean line to ax0
ax0.axhline(over_50['chol'].mean());

# add data to ax1

s=ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'],cmap='winter')

# customizing ax1
ax1.set(title='Heart diseases',xlabel='Age',ylabel='blood pressure');

# change the x and y axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
# adding legend to ax1
ax1.legend(*s.legend_elements(),title='target')


ax1.axhline(over_50['thalach'].mean());

# added title to the figure

fig.suptitle('Heart diseases Analysis',fontsize=12,fontweight='bold')
```
