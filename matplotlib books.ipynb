# subploting
#The subplot()function, in addition to subdividing the figure in different drawing areas, is used at the same time in order #focus the commands on a specific subplot.

```python
plt.subplot(211)
plt.plot(t,y1,linewidth=5.0)
plt.subplot(212)
plt.plot(t,y2,linewidth=10.0)
```

```python
#The argument of the subplot() function is composed of three integers. The first number defines how many parts the figure
#is split into vertically. The second number defines how many parts the figure is divided into horizontally.
#The third issue selects which is the current subplot on which you can direct commands.
plt.subplot(311)
plt.plot([1,2,3,4,5],[4,5,6,7,8],linewidth=3.0)
plt.title('first figure')
plt.subplot(312)
plt.plot([4,5,6],[7,8,9],linewidth=4.0)
plt.title('Second figure')
plt.subplot(313)
plt.plot([7,8,9],[10,11,12],linewidth=5.0)
plt.title('Third number')
```

# Adding a text anywhere in the program
text() function is used

# LaTeX expressions used to insert any mathematical expression in graph
start with r and enclosed by $

```python
# inserting a equation y=x^3

plt.title('My cube',color='green',fontname='Times New Roman',fontsize=30)
plt.plot([1,2,3,4,5],[1,8,27,64,125],'r-+',linewidth=4.0)
plt.xlabel('value',color='blue',fontsize=30,fontname='Times New Roman')
plt.ylabel('Cube value',color='yellow',fontsize=30)
plt.text(1,1,'first',color='violet',fontsize=50)
plt.text(4,60,'you',color='indigo',fontsize=30)
plt.text(2,100,r'$y=x^3$',fontsize=20)
```

# Adding a Grid
Adding a grid to a chart is a very simple operation: just add the grid() function, passing True as
argument

```python
plt.title('Adding grid',color='blue',fontsize=20)
plt.plot([1,2,3,4,5],'r-+',linewidth=5.0)
plt.xlabel('value',color='yellow',fontsize=20)
plt.ylabel('y value',color='green',fontsize=20)
plt.grid(True)
```

# Adding a Legend
Add a legend to your chart with the legend() function

```python
plt.title('Adding grid',color='blue',fontsize=20)
plt.plot([1,2,3,4,5],'r-+',linewidth=5.0)
plt.xlabel('value',color='yellow',fontsize=20)
plt.ylabel('y value',color='green',fontsize=20)
plt.grid(True)
plt.legend(['Numbers'])
```

```python
fig,ax=plt.subplots(figsize=(10,6))
x=np.arange(-2*np.pi,2*np.pi,0.01)
y=np.sin(3*x)/x
ax.plot(x,y)
ax.set(title='Normal distribution',xlabel='random value',ylabel='sin function')
ax.grid(True)
plt.savefig('normal distribution.jpg')
```

```python
x=np.arange(-2*np.pi,2*np.pi,0.01)
y=np.sin(3*x)/x
y1=np.sin(2*x)/x
y2=np.sin(x)/x
plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.grid(True)
plt.title('Normal distribution')
plt.xlabel('Random numbers')
plt.ylabel('sin function')
plt.legend([['sin(3*x)/x'],['sin(2*x)/x'],['sin(x)/x']])
```

```python
x=np.arange(-2*np.pi,2*np.pi,0.01)
y=np.sin(3*x)/x
y1=np.sin(2*x)/x
y2=np.sin(x)/x
plt.plot(x,y,'k--',linewidth=3)
plt.plot(x,y1,'m--',linewidth=2)
plt.plot(x,y2,color='#87a3cc',linestyle='--')
plt.grid(True)
plt.title('Normal distribution')
plt.xlabel('Random numbers')
plt.ylabel('sin function')
plt.legend([['sin(3*x)/x'],['sin(2*x)/x'],['sin(x)/x']])
```

```python
x=np.arange(-2*np.pi,2*np.pi,0.01)
y=np.sin(3*x)/x
y1=np.sin(2*x)/x
y2=np.sin(x)/x
plt.plot(x,y,'k--',linewidth=3)
plt.plot(x,y1,'m--',linewidth=2)
plt.plot(x,y2,color='#87a3cc',linestyle='--')

plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],
 [r'$-2\pi$',r'$-\pi$',r'$0$',r'$+\pi$',r'$+2\pi$'])

plt.yticks([-1,0,+1,+2,+3],[r'$-1$',r'$1$',r'$+1$',r'$+2$',r'$+3$'])


plt.title('Normal distribution')

plt.xlabel('Random numbers')

plt.ylabel('sin function')

plt.legend([['sin(3*x)/x'],['sin(2*x)/x'],['sin(x)/x']]);
```

```python
x=np.arange(-2*np.pi,2*np.pi,0.01)
y=np.sin(3*x)/x
y1=np.sin(2*x)/x
y2=np.sin(x)/x
plt.plot(x,y,'k--',linewidth=3)
plt.plot(x,y1,'m--',linewidth=2)
plt.plot(x,y2,color='#87a3cc',linestyle='--')

plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],
 [r'$-2\pi$',r'$-\pi$',r'$0$',r'$+\pi$',r'$+2\pi$'])

plt.yticks([-1,0,+1,+2,+3],[r'$-1$',r'$1$',r'$+1$',r'$+2$',r'$+3$'])


plt.title('Normal distribution')


plt.ylabel('sin function')

plt.legend([['sin(3*x)/x'],['sin(2*x)/x'],['sin(x)/x']]);
ax=plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
```

# Multiserial Bar Chart

```python
index=np.arange(5)
values1 = [5,7,3,4,6]
values2 = [6,6,4,5,7]
values3 = [5,6,5,4,6]
plt.axis([0,5,0,8])
bw = 0.3
plt.bar(index,values1,bw,color='r')
plt.bar(index+bw,values2,bw,color='g')
plt.bar(index+2*bw,values3,bw,color='b')
plt.legend(['values1','values2','values3'])
plt.title('A Multiseries Horizontal Bar Chart',fontsize=20)
```

# Multiseries Stacked Bar Charts

```python
index=np.random.randint(0,50,6)
value1=np.random.randint(0,10,6)
value2=np.random.randint(0,20,6)
value3=np.random.randint(0,30,6)
plt.bar(index,value1,color='r')
plt.bar(index,value2,color='g',bottom=value1)
plt.bar(index,value3,color='b',bottom=(value1+value2))
```

```python
import numpy as np
series1 = np.array([3,4,5,3])
series2 = np.array([1,2,2,5])
series3 = np.array([2,3,3,4])
index = np.arange(4)
plt.axis([0,4,0,15])
plt.bar(index,series1,color='r')
plt.bar(index,series2,color='b',bottom=series1)
plt.bar(index,series3,color='g',bottom=(series2+series1))
plt.xticks(index+0.4,['Jan15','Feb15','Mar15','Apr15']);
```

# Multiseries Stacked Horizonta Bar Charts

```python
import numpy as np
series1 = np.array([3,4,5,3])
series2 = np.array([1,2,2,5])
series3 = np.array([2,3,3,4])
index = np.arange(4)
plt.axis([0,15,0,4])
plt.barh(index,series1,color='r')
plt.barh(index,series2,color='b',left=series1)
plt.barh(index,series3,color='g',left=(series2+series1))
plt.yticks(index+0.4,['Jan15','Feb15','Mar15','Apr15']);
```
